```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
#print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
#print(query_url)
```

```python
city='teknaf'
response=requests.get(query_url+city).json()
print(response)
```

```python
print(response['main']['temp'])
print(response['coord']['lon'])
print(response['coord']['lat'])
```

```python

#longitude=[]
#latitude=[]
temperatureMax=[]
humidity=[]
cloudiness=[]
windSpeed=[]
output_log = "output_data/processing.log"
myFile=open(output_log,"w")
count=0

for city in cities:
    try:
        count=count+1
        response=requests.get(query_url+city).json()
        myFile.write(f'processing {count} of {len(cities)} :{city}\n')
        print(f'processing {count} of {len(cities)} :{city}')
        temperatureMax.append(response['main']['temp_max'])
        longitude.append(response['coord']['lon'])
        latitude.append(response['coord']['lat'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windSpeed.append(response['wind']['speed'])
    except:
        #myFile.write(f'exception missing city data {count} of {len(cities)} :{city}\n')
        print(f'exception missing city data {count} of {len(cities)} :{city}')
        #print(response)
```

```python
#print(temperature)

#humidity=[]
#cloudiness=[]
print(windSpeed)

```

```python
#merge arrays

city_df=pd.DataFrame(cities)
lats_df=pd.DataFrame(latitude)
lngs_df=pd.DataFrame(longitude)
temperatureMax_df=pd.DataFrame(temperatureMax)
humidity_df=pd.DataFrame(humidity)
cloudiness_df=pd.DataFrame(cloudiness)
windSpeed_df=pd.DataFrame(windSpeed)




```

```python
merged_df=pd.merge(city_df,lats_df, left_index=True, right_index=True )
merged_df=pd.merge(merged_df,lngs_df, left_index=True, right_index=True )
merged_df=pd.merge(merged_df,temperatureMax_df, left_index=True, right_index=True )
merged_df=pd.merge(merged_df,humidity_df, left_index=True, right_index=True )
merged_df=pd.merge(merged_df,cloudiness_df, left_index=True, right_index=True )
merged_df=pd.merge(merged_df,windSpeed_df, left_index=True, right_index=True )
merged_df.columns=['City','Latitude','Longitude','Temperature','Humidity','Cloudiness','Wind Speed']
merged_df.head()
```
